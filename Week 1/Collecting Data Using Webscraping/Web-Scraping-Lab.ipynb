## Extract information from the given web site

You will extract the data from the below web site: <br> 


```python
#this url contains the data you need to scrape
url = "https://cf-courses-data.s3.us.cloud-object-storage.appdomain.cloud/IBM-DA0321EN-SkillsNetwork/labs/datasets/Programming_Languages.html"
```

The data you need to scrape is the **name of the programming language** and **average annual salary**.<br> It is a good idea to open the url in your web broswer and study the contents of the web page before you start to scrape.


Import the required libraries


```python
# Your code here
from bs4 import BeautifulSoup
import requests
from openpyxl import Workbook   
```

Download the webpage at the url


```python
#your code goes here
data = requests.get(url).text
```

Create a soup object


```python
#your code goes here
soup = BeautifulSoup(data,'html5lib')
```

Scrape the `Language name` and `annual average salary`.


```python
#your code goes here
lang_dict = dict()
def get_avg_an_sal():
    for rows in soup.find_all('tr'):
        cols = rows.find_all('td')
        lang_name = cols[1].getText()
        an_av_sal = cols[3].getText()
        print(lang_name,an_av_sal)
        lang_dict[lang_name] = lang_name,an_av_sal

get_avg_an_sal()  
```

Save the scrapped data into a file named _popular-languages.csv_


```python
# your code goes here
from openpyxl import Workbook
wb = Workbook()
ws = wb.active
for key in lang_dict:
    ws.append([lang_dict[key][0],lang_dict[key][1]])
    
wb.save('popular-languages.csv')
```

```python
import xlrd
loc = ("popular-languages.csv")
wb = xlrd.open_workbook(loc)
sheet = wb.sheet_by_index(0)
print(sheet.nrows)
print(sheet.ncols)
row_num=0
while row_num < sheet.nrows:
    print(sheet.cell_value(row_num, 0), sheet.cell_value(row_num, 0))
    row_num = row_num+1
```
